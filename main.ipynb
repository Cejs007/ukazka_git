```python
vek_podle_tridy = data[["Pclass", "Age"]].groupby("Pclass").mean()
vek_podle_tridy
```

```python
# řádek po řádku, pokud je ve sloupci věk nan -> podívat do sloupce Pclass a podle hodnoty vybrat z tabulky vek_podle_tridy, pokud ne, nechat puvodni
data["Age"] = data.apply(lambda row: vek_podle_tridy.loc[row["Pclass"]].values[0] if np.isnan(row["Age"]) else row["Age"], axis=1)
```

```python
# většinou hodnota chybí, navíc hodně unikátních hodnot
data = data.drop(columns=["Cabin"])
```

```python
nejcastejsi_embarked = data["Embarked"].value_counts().idxmax()
data["Embarked"] = data["Embarked"].fillna(nejcastejsi_embarked)
```
